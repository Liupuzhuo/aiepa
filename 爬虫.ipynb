```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import time

options = webdriver.ChromeOptions()
driver = webdriver.Chrome(options=options)

url='https://image.baidu.com/search/index?tn=baiduimage&ipn=r&ct=201326592&cl=2&lm=-1&st=-1&sf=1&fmq=&pv=&ic=0&nc=1&z=&se=1&showtab=0&fb=0&width=&height=&face=0&istype=2&ie=utf-8&fm=index&pos=history&word=%E5%A4%B4%E5%83%8F'
driver.get(url)
time.sleep(30)
driver.enconding='UTF-8'
soup=BeautifulSoup(driver.page_source,'html.parser')
body=soup.find('div',attrs={'id':'wrapper'})
body=body.find('div',attrs={'id':'imgContainer'})
body=body.find('div',attrs={'id':'imgid'})

proxiex={
    "http":"http:/198.118.26.66",
    "http":"http:/196.128.27.67",
    "http":"http:/195.138.28.68",
    "http":"http:/194.148.29.69",
}

dir_file = './text/'
i=0
count=0
for txt in body.find_all('div',attrs={'class':'imgpage'}):
    txt=txt.find('ul',attrs={'class':'imglist clearfix pageNum'+str(i)})
    i+=1
    for img in txt.find_all('li',attrs={'class':'imgitem normal'}):
        img=img.find('img')
        img=img.attrs['data-imgurl']
        image=requests.get(img,proxies)
        print(img)
        string=dir_file + 'image'+ str(count) + '.jpg'
        fp = open(string,'wb')
        fp.write(image.content)
        fp.close()
        count+=1
print('图片总数:',count)
print('共:',i,'页')
```
